## Scatter Plots to Understand the Data

```python
recent_grads.plot(x ='Sample_size', y ='Median', kind = 'scatter', title = 'Sample_size vs. Median')
```

```python
recent_grads.plot(x ='Sample_size', y ='Unemployment_rate', kind = 'scatter', title = 'Sample_size vs. Unemployment_rate')
```

```python
recent_grads.plot(x ='Full_time', y ='Median', kind = 'scatter', title = 'Full_time vs. Median')
```

```python
recent_grads.plot(x ='ShareWomen', y ='Unemployment_rate', kind = 'scatter', title = 'ShareWoman vs. Unemployment_rate')
```

```python
recent_grads.plot(x ='Sample_size', y ='Unemployment_rate', kind = 'scatter', title = 'Sample_size vs. Unemployment_rate')
```

```python
recent_grads.plot(x ='Men', y ='Median', kind = 'scatter', title = 'Men vs. Median')
```

```python
recent_grads.plot(x ='Women', y ='Median', kind = 'scatter', title = 'Women vs. Median')
```

These scatter plots show that the only plot that doesn't show any kind of relation is `Unemployment_rate` and `ShareWomen`. The last plot, `Women vs. Median` shows that even though there are women earning a lot, the majority is rather around 30000 and 40000. However, the plot `Men vs Median` shows that men make a little bit more money, with incomes from around 30000 to 60000.
