# Visualising The Gender Gap In College Degrees
In this guided project I look at data from the National Center for Education Statistics and visualise the gender gap in STEM subjects.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(18, 18))

for sp in range(0,18,3):
    sp_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp_index == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2005, 8, 'Men')
    elif sp_index == 5:
        ax.text(2005, 90, 'Men')
        ax.text(2005, 5, 'Women')
        ax.tick_params(labelbottom='on')
        
for sp in range(1,16,3):
    sp_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[sp_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp_index == 0:
        ax.text(2005, 80, 'Women')
        ax.text(2005, 15, 'Men')  
    elif sp_index == 4:
        ax.tick_params(labelbottom='on')

for sp in range(2,20,3):
    sp_index = int((sp-2)/3 )
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[sp_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp_index == 0:
        ax.text(2005, 90, 'Women')
        ax.text(2005, 7, 'Men')
    elif sp_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom='on')
plt.savefig('gender_degrees.png')        
plt.show()
```

This is a rather small guided project. It aims to show off matplotlib skills. The plots show the gender gap in different subject. The subject are into three groups: STEM, arts, and others. The column on the left, middle, and right comprise the STEM, arts, and other subjects, respectively. We see that in the STEM subject, women outnumber men in Psychology and since 1990 Biology. However, the other subjects such as Engineering, Computer Science, Physical Science, and Maths and Statistics are graduated predominantly by men. In the arts subject women dominate most subjects such as Foreign Languages, English, Communication and Journalism, and Art and Performance. The subject Social Sciences and History on the other hand displays an equal share of men and women since around 2000. The last category consists of the subjects Health Professions, Public Administration, Education, Agriculture, Business and Architecture. Of these subjects, more women graduated in Health Professions, Public Administration, and Education. Agriculture and Business display an equal share of graduations of both genders in the recent years. Only Architecture has more men graduating than women, however, the trend goes towards an equal number of graduations of men and women.
